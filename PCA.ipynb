```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

```python
from sklearn import datasets
```

```python
cancer = datasets.load_breast_cancer()
```

```python
print(cancer.keys())
```

```python
print(cancer['data'])
```

```python
df = pd.DataFrame(data = cancer['data'],columns=cancer['feature_names'])
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
scaled_data
```

# PCA

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
pca
pca2=PCA(n_components=5)
pca2
```

```python
pca.fit(scaled_data)
scaled_data.shape
```

```python
pca_scaled = pca.transform(scaled_data)
```

```python
pca_scaled.shape
```

```python
plt.figure(figsize=(10,6))
plt.scatter(pca_scaled[:,0],pca_scaled[:,1],c=cancer['target'],cmap='rainbow')
```

```python
pca2_scaled = pca2.transform(scaled_data)
```

```python
pca2_scaled.shape
```

```python
pca2_scaled[:,1]
```

```python
plt.figure(figsize=(10,6))
plt.scatter(pca2_scaled[:,0],pca2_scaled[:,1],c=cancer['target'],cmap='rainbow')
```

```python
fig,(ax1,ax2) = plt.subplots(1,2,sharey=True,figsize=(10,6))

ax1.scatter(pca_scaled[:,0],pca_scaled[:,1],c=cancer['target'],cmap='rainbow')
ax1.set_title('with 2 components')

ax2.scatter(pca2_scaled[:,0],pca2_scaled[:,1],c=cancer['target'],cmap='rainbow')
ax2.set_title('with 5 components')
```
