```python
fig,axes = plt.subplots(2,2)
plt.tight_layout()

axes[0][0].plot(x,y)
axes[0][0].set_title('FIRST PLOT')
plt.tight_layout()

axes[0][1].plot(x*x,y)
axes[0][1].set_title('SECOND PLOT')
plt.tight_layout()

axes[1][0].plot(x,y*y)
axes[1][0].set_title('THIRD PLOT')
plt.tight_layout()

axes[1][1].plot(y,x)
axes[1][1].set_title('FOURTH PLOT')
plt.tight_layout()

# axes are just list of axis ie. we can iterate through them
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label = 'X Square')
ax.plot(x,x**3,label = 'X Cube')
ax.legend(loc = 0) # 0 finds the best location for legend 1->upper right   2->upper left
```

```python
# lecture 3 setting colours,line types, customizing (plot appearance)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',linewidth = 3,alpha = 0.3) # color = color of line  linewidth = width of line  alpha = transparency of line
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,linestyle = '-.') # supported line style values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,marker = 'o',markersize = 11,markerfacecolor = 'yellow',markeredgewidth = 3,markeredgecolor = 'red')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color = 'purple',lw = 2,linestyle = '-.')
ax.set_xlim(0,2)
ax.set_ylim(0,2)
```

```python
fig = plt.figure(figsize=(15,5))
x = np.arange(0,2*np.pi,0.1)
y = np.sin(x)
z = np.cos(x)
plt.plot(y,label = 'sin(x)',marker = 'X',markersize = 8,markerfacecolor = 'red')
plt.plot(z,label = 'cos(x)',marker = 'o',markersize = 8,markerfacecolor = 'purple')

plt.legend()
plt.tight_layout()
fig.savefig('sin(x),cos(x).jpg',dpi = 1000)
```
